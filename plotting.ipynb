```julia
photons_in = readdlm(pwd() * "/outputs/Test_photons_in.csv", ',', Float64);
photons_out = readdlm(pwd() * "/outputs/Test_photons_out.csv", ',', Float64);
photons_fluc = readdlm(pwd() * "/outputs/Test_photons_fluc.csv", ',', ComplexF64);

power_list = [0.01:0.005:2.5;]*1e4;

squeezing_sing = 10log10.(photons_fluc ./ photons_out);
```
